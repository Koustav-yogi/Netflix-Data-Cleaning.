```python
import pandas as pd

# Step 1: Load the Dataset
df = pd.read_csv('netflix_titles.csv')
print("Original Dataset Shape:", df.shape)
print(df.head())

# Step 2: Check for Missing Values
print("\nMissing Values:\n", df.isnull().sum())

# Fill missing 'director' and 'country' with 'Unknown'
df['director'] = df['director'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')

# For 'date_added', convert to datetime and fill any NaT
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
print("\nMissing after conversion:\n", df.isnull().sum())
df['date_added'] = df['date_added'].fillna(pd.Timestamp('2000-01-01'))  # or drop rows

# Step 3: Remove Duplicate Rows
print("\nDuplicate Rows:", df.duplicated().sum())
df = df.drop_duplicates()

# Step 4: Standardize Text Values
df['country'] = df['country'].str.strip().str.title()
df['type'] = df['type'].str.strip().str.title()

# Step 5: Clean Column Names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
print("\nUpdated Columns:\n", df.columns)

# Step 6: Check and Fix Data Types
print("\nData Types:\n", df.dtypes)

# Step 7: Export Cleaned Dataset
df.to_csv('cleaned_netflix_titles.csv', index=False)
print("\nCleaned dataset saved as 'cleaned_netflix_titles.csv'.")

# Step 8: Final Shape
print("\nFinal Dataset Shape:", df.shape)
```
